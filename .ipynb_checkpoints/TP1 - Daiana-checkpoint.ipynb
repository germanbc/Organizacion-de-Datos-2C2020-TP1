```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import cycle, islice
import datetime as datetime
import glob

%matplotlib inline

plt.style.use('default') # haciendo los graficos un poco mas bonitos en matplotlib
#plt.rcParams['figure.figsize'] = (20, 10)

sns.set(style="whitegrid") # seteando tipo de grid en seaborn

pd.options.display.float_format = '{:20,.2f}'.format # suprimimos la notacion cientifica en los outputs

#import warnings
#warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (15, 5)
```

```python
eci = pd.read_csv('Entrenamieto_ECI_2020.csv')
```

```python
aprobadas = eci[eci['Pricing, Delivery_Terms_Approved'] > 0]
aprobadasPorRegion = aprobadas['Region'].value_counts()\
        .to_frame().reset_index()\
        .rename(columns = {'index':'Región','Region':'Aprobadas'})
aprobadasPorRegion
```

```python
aprobadasPorRegion.plot(kind='bar', x='Región', y='Aprobadas',\
                 width=0.7, figsize=(11,5.5), color='blue',\
                 edgecolor="lightblue", linewidth=4, rot=50)
g=plt.gca()
g.set_title("¿Cuántas oportunidades se aprobaron por región?",\
            fontsize=20);
g.set_xlabel("Región", fontsize=18);
g.set_ylabel("Aprobadas", fontsize=18);
```

```python
desaprobadas = eci[eci['Pricing, Delivery_Terms_Approved'] < 1]
desaprobadasPorRegion = desaprobadas['Region'].value_counts()\
    .to_frame().reset_index()\
    .rename(columns = {'index':'Región','Region':'Desaprobadas'})
desaprobadasPorRegion
```

```python
desaprobadasPorRegion.plot(kind='bar', x='Región',y='Desaprobadas',\
                 width=0.7, figsize=(11,5.5), color='red',\
                 edgecolor="pink", linewidth=4, rot=50)
g=plt.gca()
g.set_title("¿Cuántas oportunidades se desaprobaron por región?",\
            fontsize=20);
g.set_xlabel("Región", fontsize=18);
g.set_ylabel("Desaprobadas", fontsize=18);
```

```python
ganadas = eci[eci['Stage'] == 'Closed Won']
ganadasPorTerritorio = ganadas['Territory'].value_counts()\
        .to_frame().reset_index()\
        .rename(columns = {'index':'Territorio','Territory':'Closed Won'})
ganadasPorTerritorio
```

```python
ganadasPorTerritorio.head(10).plot(kind='bar', x='Territorio', y='Closed Won',\
                 width=0.7, figsize=(11,5.5), color='yellow',\
                 edgecolor="lightgray", linewidth=4, rot=50)
g=plt.gca()
g.set_title("¿Oportunidades Closed Won por territorio? (top 10)",\
            fontsize=20);
g.set_xlabel("Territorio", fontsize=18);
g.set_ylabel("Ops Closed Won", fontsize=18);
```

```python
perdidas = eci[eci['Stage'] == 'Closed Lost']
perdidasPorTerritorio = perdidas['Territory'].value_counts()\
        .to_frame().reset_index()\
        .rename(columns = {'index':'Territorio','Territory':'Closed Lost'})
perdidasPorTerritorio
```

```python
perdidasPorTerritorio.head(10).plot(kind='bar', x='Territorio', y='Closed Lost',\
                 width=0.7, figsize=(11,5.5), color='orange',\
                 edgecolor="lightblue", linewidth=4, rot=50)
g=plt.gca()
g.set_title("¿Oportunidades Closed Lost por territorio? (top 10)",\
            fontsize=20);
g.set_xlabel("Territorio", fontsize=18);
g.set_ylabel("Ops Closed Lost", fontsize=18);
```
